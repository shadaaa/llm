```python
for data in[legit,attack,traffic]:
  data['hour'] = data['@timestamp'].dt.hour
  data['day'] = data['@timestamp'].dt.dayofweek
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
features = ['method','ssl-cipher','ssl-protocol','domain']
encoder = LabelEncoder()
for data in [legit,attack,traffic]:
  for col in features:
    if data[col].dtype == 'object':

      data[col] = encoder.fit_transform(data[col])
```

```python
training_features = ['body_bytes_sent', 'connection_requests', 'connection_time',
    'geoiplatt', 'geoiplong', 'request_length', 'request_time',
    'ssl-cipher', 'ssl-protocol', 'method', 'hour', 'day']
scaler = StandardScaler()
legit_features = scaler.fit_transform(legit[features])
attack_features = scaler.transform(attack[features])
traffic_features = scaler.transform(traffic[features])
```

```python
model = IsolationForest(contamination=0.1,random_state=42)
model.fit(legit_features)
```

```python
import numpy as np
def prediction(data,label):
  data_predict = data[features]
  predictions = model.predict(data_predict)
  predictions = np.where(predictions == -1,1,0)
  print(f"{label}")
  print(classification_report([1] * len(predictions),predictions))
  print(confusion_matrix([1] * len(predictions),predictions))
  return predictions
predicts = prediction(traffic,"Traffic Logs")
attack_preds = prediction(attack,"Attack Logs")
```
